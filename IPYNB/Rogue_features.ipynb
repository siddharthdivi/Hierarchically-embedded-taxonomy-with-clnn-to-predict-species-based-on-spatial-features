```python
archive = zipfile.ZipFile('data.zip', 'r')
data = pd.read_csv(archive.open("data.tsv"),delimiter='\t',header=None)
data.columns = ['queryID','query','response','label','labelID']
data.isnull().values.any()
```

```python
def basic_cleaning(df,column):
  #Split a-b into a and b
  df[column] = df[column].str.replace('-',' ')
  print(column+": - Replaced")
  #Replace all numbers
  #df[column] = df[column].str.replace('[0-9]+','')
  #print(column+": Numbers Replaced")
  #Convert to lowercase
  df[column] = df[column].apply(lambda x: " ".join(x.lower() for x in x.split()))
  print(column+": Converted to lowercase")
  #Remove punctuations
  df[column] = df[column].str.replace('[^\w\s]','')
  print(column+": Removed punctions ")
  return df[column]
```

```python
data['query'] = basic_cleaning(data,'query')
data['response'] = basic_cleaning(data,'response')
```

```python
from allennlp.predictors.predictor import Predictor
predictor = Predictor.from_path("https://s3-us-west-2.amazonaws.com/allennlp/models/decomposable-attention-elmo-2018.02.19.tar.gz")
```

```python
exp = data.copy(deep=True)#pr.iloc[:100].copy(deep = True)
exp['Entailment'] = 1
exp['Contradiction'] = 1
exp['Neutral'] = 1
for i in exp.index:
  l = predictor.predict(hypothesis=exp['query'][i],premise=exp['response'][i])['label_probs']
  exp['Entailment'][i] = l[0]*100
  exp['Contradiction'][i] = l[1]*100
  exp['Neutral'][i] = l[2]*100
  if(i%25==0):
    print(i)
  if(i%100 == 0):
    exp.to_csv('Allen_Data.csv',index=False)
  
```
