```python
archive = zipfile.ZipFile('data.zip', 'r')
data = pd.read_csv(archive.open("data.tsv"),delimiter='\t',header=None)
data.columns = ['queryID','query','response','label','labelID']
data.isnull().values.any()
```

```python
def basic_cleaning(df,column):
    #Split a-b into a and b
    df[column] = df[column].str.replace('-',' ')
    print(column+": - Replaced")
    #Replace all numbers
    df[column] = df[column].str.replace('[0-9]+','')
    print(column+": Numbers Replaced")
    #Convert to lowercase
    df[column] = df[column].apply(lambda x: " ".join(x.lower() for x in x.split()))
    print(column+": Converted to lowercase")
    #Remove punctuations
    df[column] = df[column].str.replace('[^\w\s]','')
    print(column+": Removed punctions")
    return df[column]
```

```python
data['query'] = basic_cleaning(data,'query')
data['response'] = basic_cleaning(data,'response')
```

```python
from InferSent.models import InferSent
import torch
V = 2
MODEL_PATH = './InferSent/encoder/infersent%s.pkl' % V
params_model = {'bsize': 64, 'word_emb_dim': 300, 'enc_lstm_dim': 2048,
                'pool_type': 'max', 'dpout_model': 0.0, 'version': V}
infersent = InferSent(params_model)
infersent.load_state_dict(torch.load(MODEL_PATH))
```

```python
W2V_PATH = './InferSent/dataset/fastText/crawl-300d-2M.vec'
infersent.set_w2v_path(W2V_PATH)
```

```python
infersent = infersent.cuda()
```

```python
infersent.build_vocab(data['query'], tokenize=True)
```

```python
infersent.update_vocab(data['response'], tokenize=True)
```

```python
size = 10000
for i in range(0,data.shape[0],size):
    q_embeddings = infersent.encode(data['query'][i:i+size], tokenize=True, verbose=True)
    r_embeddings = infersent.encode(data['response'][i:i+size], tokenize=True, verbose=True)
    pd.DataFrame(q_embeddings).to_csv("query_embeddings_"+str(i)+".csv",sep=',',index=None)
    pd.DataFrame(r_embeddings).to_csv("response_embeddings_"+str(i)+".csv",sep=',',index=None)
    del q_embeddings
    del r_embeddings
```
