# Analyze the hierarchy

```python
# Consider only the class labels required to predict
d1 = data[['class','order','family','genus','species_glc_id']]
```

```python
d1
```

```python
# find all unique classes, in the given dataset
uniqueClasses = d1['class'].unique()
a = {} 
a = defaultdict(lambda:0,a)
classDict = {}
for cls in uniqueClasses:
    # For each unique class, find the orders present in them
    clssum=0;
    i=0;
    orderDict = {}
    uniqueOrder = d1[d1['class']==cls]['order'].unique()
    for ordr in uniqueOrder:
        # For each unique order, find the families present in them
        osum=0
        i+=1
        j=0
        familyDict = {}
        uniqueFamily = d1[d1['order']==ordr]['family'].unique()
        for fam in uniqueFamily:
            # For each unique family, find the genus present in them
            j+=1
            fsum=0
            genusDict = {}
            uniqueGenus = d1[d1['family']==fam]['genus'].unique()
            for gen in uniqueGenus:
                # For each unique genome, find the species present in them
                uniqueSpecies = d1[d1['genus']==gen]['species_glc_id'].unique()
                # store the count of unique species 
                # (no. of unique species, no. of times this count appears in the hierarchies seen)
                clssum+=len(uniqueSpecies);
                osum+=len(uniqueSpecies);
                fsum+=len(uniqueSpecies);
                a[len(uniqueSpecies)] += 1
                genusDict[gen] = uniqueSpecies
            familyDict[fam] = genusDict
            #3.Classname->order_name->family_name->no_of_genus->no_of_species
            print("\n",cls,"->",i,"."+ordr,"->",j,"."+fam,"->",len(uniqueGenus),"->",fsum)
        orderDict[ordr] = familyDict
        #2.Classname->order_name->no_of_families->no_of_species
        #print("\n",cls,"->",i,"."+ordr,"->",len(uniqueFamily),"->",osum)
    classDict[cls] = orderDict
    #1.Classname->no_of_orders->no_of_species
    #print("\n",cls,"->",len(uniqueOrder),"->",clssum)
    
```
